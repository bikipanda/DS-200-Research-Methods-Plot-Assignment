```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('C:/Users/pandaonymous/Desktop/New/1.csv')
```

```python
columns_ = np.array(data.columns)
#print(columns_)
male_pos = [columns_[1],columns_[3],columns_[5],columns_[7],columns_[9],columns_[11]]
print("male columns : ",male_pos)
female_pos = [columns_[2],columns_[4],columns_[6],columns_[8],columns_[10],columns_[12]]
print("female columns : " , female_pos)
```

```python
male_data = data.loc[36,male_pos]
female_data = data.loc[36,female_pos]

male_plot_points = male_data.to_numpy()
female_plot_points = female_data.to_numpy()
x_axis = ['<18','18-25','25-35','35-45','45-60','>=60']
```

```python
bar1 = np.arange(len(x_axis))
bar2 = bar1+0.25
plt.bar(bar1,male_plot_points, color = 'b', width = 0.25,label ="Men")
plt.bar(bar2,female_plot_points, color = 'g', width = 0.25, label ="Women")
plt.xlabel('Age in Years')
plt.ylabel('Number of Fatal road accident')
plt.title('Fatal accidents in 2017 classified by Age and Sex')
plt.xticks(bar1+0.12,x_axis)
plt.legend()
#for i, male_plot_points in zip(bar1, male_plot_points):
#    plt.text(bar1[i] - 0.25, male_plot_points + 0.5, str(male_plot_points))
#for j, female_plot_points in zip(bar2, female_plot_points):
#    plt.text(bar2[j] - 0.25, female_plot_points + 0.5, str(female_plot_points))

for i, v in enumerate(male_plot_points):
    plt.text(bar1[i] - 0.25, v + 0.1, str(v))
for j, z in enumerate(female_plot_points):
    plt.text(bar2[j] - 0.1 , z + 0.1, str(z))
    
plt.savefig('C:/Users/pandaonymous/Desktop/Fatal Accidents by Age and sex in 2017.jpg',dpi = 400)
plt.show()
```
